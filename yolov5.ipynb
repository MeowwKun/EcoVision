```python
import cv2
import os
from pathlib import Path

def resize_to_640_preserve_originals(base_path, subdirs, target_size=(640, 640)):
    for subdir in subdirs:
        original_folder = Path(base_path) / subdir
        resized_folder = original_folder.parent / f"{original_folder.name}_640"
        resized_folder.mkdir(parents=True, exist_ok=True)

        for img_path in original_folder.glob("*.*"):
            if img_path.suffix.lower() not in ['.jpg', '.jpeg', '.png']:
                continue

            img = cv2.imread(str(img_path))
            if img is None:
                print(f"Skipping unreadable image: {img_path}")
                continue

            resized = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)

            # Save as .jpg in the new folder
            save_path = resized_folder / (img_path.stem + ".jpg")
            cv2.imwrite(str(save_path), resized)
            print(f"Saved resized: {save_path}")

base_path = "/content/drive/MyDrive/Colab Notebooks/PlasticBottleData"
subdirs = ["train/images", "valid/images", "test/images"]

resize_to_640_preserve_originals(base_path, subdirs)
```

```python
import os

base_path = "/content/drive/MyDrive/Colab Notebooks/PlasticBottleData"
for split in ["train", "valid"]:
    image_dir = os.path.join(base_path, split, "images_640")
    label_dir = os.path.join(base_path, split, "labels_640")

    image_files = [f for f in os.listdir(image_dir) if f.endswith(".jpg")]
    renamed = 0

    for image_file in image_files:
        image_name = os.path.splitext(image_file)[0]
        label_file = image_name + ".txt"

        # Try to find any label file that has the same hash portion
        image_hash = image_file.split("rf.")[-1].split(".")[0]
        label_candidates = [f for f in os.listdir(label_dir) if image_hash in f]

        if label_candidates:
            old_label_path = os.path.join(label_dir, label_candidates[0])
            new_label_path = os.path.join(label_dir, label_file)
            os.rename(old_label_path, new_label_path)
            renamed += 1

    print(f"{split.upper()}: Renamed {renamed} label files.")
```

```python
from ultralytics import YOLO
import cv2
import torch
import matplotlib.pyplot as plt

# Load the trained YOLO model
model = YOLO("/content/runs/detect/train/weights/yolo11n.pt")  # Update path if needed

# Path to the image you want to detect objects on
image_path = "/content/drive/MyDrive/bottle.jpeg"  # Update this with your image path

# Perform inference (detection)
results = model(image_path)

# Read the original image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct visualization

# Process results
for r in results:
    if len(r.boxes) > 0:  # Ensure there are detections
        # Get confidence scores and find the highest one
        confs = r.boxes.conf  # Confidence scores
        max_conf_index = torch.argmax(confs).item()  # Index of highest confidence detection
        best_box = r.boxes[max_conf_index]  # Get the bounding box with the highest confidence

        # Extract bounding box coordinates
        x1, y1, x2, y2 = map(int, best_box.xyxy[0].tolist())
        label = f"{r.names[int(best_box.cls)]}: {best_box.conf.item():.2f}"

        # Draw the highest confidence detection on the image
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 3)  # Green box
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        # Display the image with the single best detection
        plt.figure(figsize=(10, 6))
        plt.imshow(image)
        plt.axis("off")
        plt.show()
    else:
        print("No detections found.")
```
